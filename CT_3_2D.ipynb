```python
from math import *
import numpy as np
from numpy.linalg import eig
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
mu, k = 3, 45
```

```python
kp, kd = 100, 2
```

```python
#no control
def state_array(x, t):
    A = np.array([[10, 3], [5, -5]]) #state matrix
    return A.dot(x)
```

```python
x0 = np.random.rand(2)
time = np.linspace(0, 1, 1000)
```

```python
solution = {"before control": odeint(state_array, x0, time)}
```

```python
#plot without control
plt.plot(time, solution["before control"])
plt.subplot(1, 1, 1)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```

```python
#using control
def controlled_MIMO(x,t):
    B = np.array([[1, 0], [0,1]]) #input matrix
    x_desired = np.array([2,1])
    x_dot_desired = np.array([1,1])
    
    error = x_desired - x
    error_dot = x_dot_desired - array(x, t) 
    
    u = kp* error + kd * error_dot
    return state_array(x,t) + B.dot(u)
```

```python
solution_c = {"controlled": odeint(controlled_MIMO, x0, time)}
```

```python
#plot with control
plt.plot(time, solution_c["controlled"])
plt.subplot(1, 1, 1)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```
